**To predict salary based on Years of Experience using Simple Linear Regression**

```python
import pandas as pd
import numpy as np

df=pd.read_csv('Salary_data.csv')
df.head()
```

```python
#Without libraries
import matplotlib.pyplot as plt

X=df['YearsExperience']
y=df['Salary']
N=np.size(X)
xsum=np.sum(X)
ysum=np.sum(y)
x2sum=np.sum(X**2)
y2sum=np.sum(y**2)
xysum=np.sum(X*y)

m=(N * xysum - xsum * ysum) / (N * x2sum - xsum**2)
c=(ysum - m * xsum) / N

print(f"Slope = {m:.2f}, Intercept = {c:.2f}")

def predict(x):
    return m*x+c

y_predict=np.array([predict(x) for x in X])

df['Predicted Salary']=y_predict

plt.scatter(X,y,color='red')
plt.plot(X, y_predict, color='blue', label='Predictions')
plt.xlabel('Years of Experience')
plt.ylabel('Salary')
plt.title('Salary vs Years of Experience')
plt.legend()
plt.show()
```

```python
#With libraries
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error

X=np.array(df['YearsExperience']).reshape(-1,1)
y=np.array(df['Salary']).reshape(-1,1)

Rmodel=LinearRegression()

Rmodel.fit(X,y)

slope=Rmodel.coef_
intercept=Rmodel.intercept_
print(f"Slope = {slope}, Intercept = {intercept}\n")

y_predict=Rmodel.predict(X)

rmse=np.sqrt(mean_squared_error(y,y_predict))
y_mean=np.mean(y)
sse=np.sum((y-y_predict)**2)
ssr=np.sum((y_predict-y_mean)**2)
sst=np.sum((y-y_mean)**2)
print(f"RMSE = {rmse:.2f}\nSSE = {sse:.2f}\nSSR = {ssr:.2f}\nSST = {sst:.2f}")

plt.scatter(X, y, color='red')
plt.plot(X, y_predict, color='blue', label='Predictions')
plt.xlabel('Years of Experience')
plt.ylabel('Salary')
plt.title('Salary vs Years of Experience')
plt.legend()
plt.show()
```

**Implement simple linear regression to predict the price of car based on mileage**

```python
df=pd.read_excel('CarPrice.xlsx')
df.head()
```

```python
#Without libraries
X=np.array(df['Mileage']).reshape(-1,1)
y=np.array(df['Price']).reshape(-1,1)

N=np.size(X)
xsum=np.sum(X)
ysum=np.sum(y)
x2sum=np.sum(X**2)
y2sum=np.sum(y**2)
xysum=np.sum(X*y)

m=(N * xysum - xsum * ysum) / (N * x2sum - xsum**2)
c=(ysum - m * xsum) / N

print(f"Slope = {m:.2f}, Intercept = {c:.2f}")

def predict(x):
    return m*x+c

y_predict=np.array([predict(x) for x in X])

df['Predicted Price']=y_predict

plt.scatter(X,y,color='red')
plt.plot(X, y_predict, color='blue', label='Predictions')
plt.xlabel('Mileage')
plt.ylabel('Price')
plt.title('Price vs Mileage')
plt.legend()
plt.show()
```

```python
#With libraries
rmodel=LinearRegression()
rmodel.fit(X,y)

slope=rmodel.coef_
intercept=rmodel.intercept_
print(f"Slope = {slope}, Intercept = {intercept}\n")

y_predict=rmodel.predict(X)

rmse=np.sqrt(mean_squared_error(y,y_predict))
y_mean=np.mean(y)
sse=np.sum((y-y_predict)**2)
ssr=np.sum((y_predict-y_mean)**2)
sst=np.sum((y-y_mean)**2)
print(f"RMSE = {rmse:.2f}\nSSE = {sse:.2f}\nSSR = {ssr:.2f}\nSST = {sst:.2f}")

plt.scatter(X,y,color='red')
plt.plot(X,y_predict,color="green")
plt.xlabel('Mileage')
plt.ylabel('Price')
plt.title('Price vs Mileage')
plt.show()
```

**To predict the sales of car in units based on mileage and price using multiple linear regression**

```python
df=pd.read_csv('carprices.csv')
df.head()
```

```python
X=df[['Mileage','Price']]
y=df['Sales']

r_model=LinearRegression()

r_model.fit(X,y)

slope=r_model.coef_
intercept=r_model.intercept_
print(f"Slope = {slope}, Intercept = {intercept}\n")

y_predict=r_model.predict(X)

rmse=np.sqrt(mean_squared_error(y,y_predict))
y_mean=np.mean(y)
sse=np.sum((y-y_predict)**2)
ssr=np.sum((y_predict-y_mean)**2)
sst=np.sum((y-y_mean)**2)
print(f"RMSE = {rmse:.2f}\nSSE = {sse:.2f}\nSSR = {ssr:.2f}\nSST = {sst:.2f}")

plt.figure(figsize=(8, 5))
plt.scatter(range(len(y)), y, color='red', label='Actual Sales', alpha=0.7)
plt.scatter(range(len(y_predict)), y_predict, color='blue', label='Predicted Sales', alpha=0.7)
plt.plot(range(len(y)), y, color='red', alpha=0.5)
plt.plot(range(len(y_predict)), y_predict, color='blue', alpha=0.5)

plt.xlabel('Data Points')
plt.ylabel('Sales')
plt.title('Actual vs. Predicted Sales')
plt.legend()
plt.show()
```
